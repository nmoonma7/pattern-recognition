```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from skimage import io, color
from skimage import exposure
```

```python
file_image = 'cau.jpg'

im_color = io.imread(file_image)
im_gray  = color.rgb2gray(im_color)

plt.plot(im_gray)
plt.imshow(im_gray, cmap='gray')
plt.axis('off')

plt.show()
```

# $ u^{(t+1)}=u^{(t)}+ \Delta t \alpha \Delta u^{(t)}$
# $ \Delta t=1/4 , \alpha = 1/4 $

```python

for i in range(8):
    
    t= 1/4
    a= 1/4
    
    j= 2**i
    u_curr= im_gray
    u_next= u_curr
    
    print('{}th iteration'.format(j))
    for i in range(j):
        
        pad=np.pad(u_next, 1, mode='constant')
        up=np.roll(pad, 1, axis=0)
        down=np.roll(pad, -1, axis=0)
        left=np.roll(pad, 1, axis=1)
        right=np.roll(pad, -1, axis=1)
        center=-(pad*4)
        
        result= up+down+left+right+center
        
        num_row,num_col=result.shape
        final_result=t*a*result[1:num_row-1,1:num_col-1]
        u_next= final_result
        
    plt.plot(u_next)
    plt.imshow(u_next, cmap='gray')
    plt.axis('off')
    plt.show()
        
    
```
